## 2.1

Find the number of characters, words and sentences for the mentioned documents. (Hint: use .words(), .raw(), .sent() for extracting counts)

```python
##
##

import nltk
nltk.download('punkt')
```

```python
##
## Roosevelt's speech characteristics

roosevelt_1941_text = inaugural.raw('1941-Roosevelt.txt')
roosevelt_1941_words = inaugural.words('1941-Roosevelt.txt')
roosevelt_1941_sents = inaugural.sents('1941-Roosevelt.txt')

# Finding the number of characters, words, and sentences
num_chars_1941 = len(roosevelt_1941_text)
num_words_1941 = len(roosevelt_1941_words)
num_sents_1941 = len(roosevelt_1941_sents)

print(f"Roosevelt 1941: Characters={num_chars_1941}, Words={num_words_1941}, Sentences={num_sents_1941}")
```

```python
##
## Kennedy's speech characteristics

kennedy_1961_text = inaugural.raw('1961-Kennedy.txt')
kennedy_1961_words = inaugural.words('1961-Kennedy.txt')
kennedy_1961_sents = inaugural.sents('1961-Kennedy.txt')

# Finding the number of characters, words, and sentences
num_chars_1961 = len(kennedy_1961_text)
num_words_1961 = len(kennedy_1961_words)
num_sents_1961 = len(kennedy_1961_sents)

print(f"Kennedy 1961: Characters={num_chars_1961}, Words={num_words_1961}, Sentences={num_sents_1961}")
```

```python
##
## Nixon's speech characteristics

nixon_1973_text = inaugural.raw('1973-Nixon.txt')
nixon_1973_words = inaugural.words('1973-Nixon.txt')
nixon_1973_sents = inaugural.sents('1973-Nixon.txt')

# Finding the number of characters, words, and sentences
num_chars_1973 = len(nixon_1973_text)
num_words_1973 = len(nixon_1973_words)
num_sents_1973 = len(nixon_1973_sents)

print(f"Nixon 1973: Characters={num_chars_1973}, Words={num_words_1973}, Sentences={num_sents_1973}")
```

## 2.2

Remove all the stopwords from the three speeches. Show the word count before and after the removal of stopwords. Show a sample sentence after the removal of stopwords.

```python
##
##

import nltk
nltk.download('stopwords')
```

```python
##
## function to remove stopwords

from nltk.corpus import stopwords
import string

## List of stopwords
stop_words = set(stopwords.words('english'))

## Function to remove stopwords and punctuation
def remove_stopwords(text_words):
    ##return [word.lower() for word in text_words if word.lower() not in stop_words and word not in string.punctuation]
    return [word.lower() for word in text_words if word.lower() not in stop_words and word.isalpha()]
```

```python
##
## removing stopwords for Roosevelt's Speech

roosevelt_1941_filtered = remove_stopwords(roosevelt_1941_words)

## You can print the length to check how many words are left after removing stopwords
print(f"Roosevelt 1941: Words before removing stopwords = {len(roosevelt_1941_words)}")
print(f"Roosevelt 1941: Words after removing stopwords = {len(roosevelt_1941_filtered)}")

## sample
print(roosevelt_1941_filtered)
```

```python
##
## removing stopwords for Kennedy's Speech

kennedy_1961_filtered = remove_stopwords(kennedy_1961_words)

# You can print the length to check how many words are left after removing stopwords
print(f"Kennedy 1961: Words before removing stopwords = {len(kennedy_1961_words)}")
print(f"Kennedy 1961: Words after removing stopwords = {len(kennedy_1961_filtered)}")

## sample
print(kennedy_1961_filtered)
```

```python
##
## removing stopwords for Nixon's Speech

nixon_1973_filtered = remove_stopwords(nixon_1973_words)

# You can print the length to check how many words are left after removing stopwords
print(f"Nixon 1973: Words before removing stopwords = {len(nixon_1973_words)}")
print(f"Nixon 1973: Words after removing stopwords = {len(nixon_1973_filtered)}")

## sample
print(nixon_1973_filtered)
```

## 2.3

Which word occurs the most number of times in his inaugural address for each president? Mention the top three words. (after removing the stopwords)

```python
##
##

from nltk.probability import FreqDist
```

```python
##
## Roosevelt's most frequent words


## Create a frequency distribution object
roosevelt_1941_freq_dist = FreqDist(roosevelt_1941_filtered)

## Get the top 3 most common words
roosevelt_1941_most_common = roosevelt_1941_freq_dist.most_common(3)

print(f"Roosevelt 1941: Top 3 most frequent words = {roosevelt_1941_most_common}")
```

```python
##
## Kennedy's most frequent words


## Create a frequency distribution object
kennedy_1961_freq_dist = FreqDist(kennedy_1961_filtered)

## Get the top 3 most common words
kennedy_1961_most_common = kennedy_1961_freq_dist.most_common(3)

print(f"Kennedy 1961: Top 3 most frequent words = {kennedy_1961_most_common}")
```

```python
##
## Nixon's most frequent words


## Create a frequency distribution object
nixon_1973_freq_dist = FreqDist(nixon_1973_filtered)

## Get the top 3 most common words
nixon_1973_most_common = nixon_1973_freq_dist.most_common(3)

print(f"Nixon 1973: Top 3 most frequent words = {nixon_1973_most_common}")
```

## 2.4

Plot the word cloud of each of the three speeches. (after removing the stopwords)

```python
##
##

from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
##
## Roosevelt's word cloud

## Generate the word cloud
roosevelt_1941_wordcloud = WordCloud(width=800, height=400,
                                    colormap="Purples",
                                    max_font_size=200,
                                    min_font_size=10,
                                    background_color="white").generate(' '.join(roosevelt_1941_filtered))


## Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(roosevelt_1941_wordcloud, interpolation = 'lanczos')
plt.axis("off")
plt.title("Roosevelt 1941 Speech Word Cloud")
plt.show()
```
